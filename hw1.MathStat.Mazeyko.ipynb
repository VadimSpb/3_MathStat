### Задание 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. 
### <div> a) Найти вероятность того, что все карты – крести. 

В колоде из 52 карт имеется 13  карт каждой масти. Таким образом, для каждой вынутой карты вероятность того, что выпадет крести составляет: 

$$\frac{количество\:крести\:в\:колоде-число\:попыток}{количество\:карт\:в\:колоде-число\:попыток}$$

Так как событие совместное, последовательно умножаем вероятности выпадания крести для каждой попытки: 

$$\frac{13-0}{52-0}*\frac{13-1}{52-1}*\frac{13-2}{52-2}*\frac{13-3}{52-3} = 
\frac{13}{52}*\frac{12}{51}*\frac{11}{50}*\frac{10}{49} = $$

$$= \frac{52821}{20000000} ≈ 0.0026410564225690276$$

### <div> б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Решим эту задачу от обратного. 
Итак, общее число возможных комбинаций:

$$ C_{n}^{m} = C_{количество\:возможных\:комбинаций}^{количество\:карт\:в\:колоде} =  C_{52}^{4} =  \frac{52!}{(52-4)! * 4!} = 270\:725$$

Теперь найдём количество сочетаний БЕЗ тузов:  

$$ C_{4}^{52-4} = C_{4}^{48} =  \frac{48!}{(48-4)!*4!} = 194\:580$$

Теперь вычтем из общего количества комбинаций количество комбинаций, не удовлетворяющих условию:

$$ С_{есть\:хотя\:6ы\:один\:туз} = C_{общ} - С_{нет\:тузов} = 270\:725 - 194\:580  = 76\:145$$

Таким образом, вероятность P события А равняется отношению количества комбинаций, удовлетворяющих условию, от общего числа комбинаций:
$$ P(A) = \frac{76\:145}{270\:725} ≈ 0,28 $$

### Задание 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Здесь задача на размещение. Воспользовался <a href='https://narfu.ru/university/library/books/1013.pdf'>этим</a> очень хорошим конспектом. 
Число возможных комбинаций по формуле:
$$ C_{n}^{m} = \frac{n!}{(n-m)!*n!} = C_{3}^{10} =\frac{10!}{(10-3)!*3!} = 120 $$

Правильной может быть только одна комбинация из трёх кнопок. Таким образом,  вероятность наступления события составляет: 

$$ P(A) = \frac{1}{120} ≈ 0,008 $$

### Задача 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Как и в задаче 1, здесь используем формулу умножения вероятностей благополучных событий:

$$ P(A) =\frac{количество\:окрашенных\:деталей-число\:попыток}{количество\:всех\:деталей-число\:попыток}$$

Таким образом, вероятность успешного события равна:

$$ P(A) = \frac{9-0}{15-0}*\frac{9-1}{15-1}*\frac{9-2}{15-2} = 
\frac{9}{15}*\frac{8}{14}*\frac{7}{13} ≈ 0,18$$

### Задача 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Также решим через умножение вероятностей. 
Вероятность первого благополучного исхода: 

$$ \frac{2}{100} $$

Вероятность второго благополучного исхода: 

$$ \frac{1}{99} $$

Таким образом, вероятность вытянуть оба счастливых билета:

$$ P(A)  = \frac{2}{100} * \frac{1}{99} ≈ 0,0002 $$
